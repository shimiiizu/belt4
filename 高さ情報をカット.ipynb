```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 事前にcsvの保存方式をカンマ区切りに変える必要があるので注意
df = pd.read_csv('./No1_10_height_s.csv', encoding='shift_jis',delimiter=',',skiprows=51)
```

```python
# [top : bottom, left : right]   1024/64=16, 768/64=12
# サンプル1の切り出し、保存

for i in range(12):
  for j in range(16):
    print(i,j)
    df_cut = df.iloc[64*i : 64*(i+1), 64*j: 64*(j+1)]
    df_cut.to_csv("height_cut_test"+str(i)+"_"+str(j)+".csv" ,header=False,index = False)
```
